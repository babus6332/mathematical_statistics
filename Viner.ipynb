## Винеровский процесс


Рассмотрим "темродинамический" предел симметричного случайного блуждания:

$(n, S_n), n = 0, 1, 2, ...$

где $S_0 = 0, S_n = \sum ^n_{n=1}   \xi_i, n \geqslant 1$
$\xi_i$ - независимые случайно распределенные с. в.

$\xi_i$ = +1, c p = 0.5 и
-1, с p = 0.5

```python
import numpy as np
```

```python
#реализуем S(n) рекуррсивно
def S(N):
    if (N==0):
        q = 0
    else:
        q = S(N-1) + np.random.choice ([-1,1], p = [0.5, 0.5]) 
        #независимые случайно распределенные с. в. \xi_i = +1, c p = 0.5 и -1, с p = 0.5
    return q
```

```python
N = 100 #число точек в выборке
a = [0]*N
for i in range(N):
    a[i] = i/N
b = [0]*N

for i in range (N):
    b[i] = S(i)/np.sqrt(N)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot (a,b)
```

```python
fig, ax = plt.subplots(figsize = [20, 10])
for i in range (3):
    N = (i+1)*50
    a = np.linspace (1, N, N)
    b = [0]*N
    for i in range (N):
        b[i] = S(i)
    a = a/N
    b = b / np.sqrt (N)
    plt.plot (a,b)
ax.grid('True')
```
